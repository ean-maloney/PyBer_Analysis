```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
#Set x-axis to months
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

#Set y-axis to total fare in $ for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create new plot andgive label to legend
plt.plot(x_axis, y_axis, label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set y limit
plt.ylim(0,45)

#Create title
plt.title("PyBer fare by Month")
#Add legend
plt.legend()
```

```python
#create plot
plt.plot(x_axis, y_axis, marker = "*", color = "b", linewidth = 2, label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.ylim(0,45)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis, label = "Boston", color = "g", marker = "D", linewidth = 2)
ax.legend()
ax.set_xlabel("Month")
ax.set_ylabel("Fare ($)")
ax.grid()
ax.set_ylim(0, 45)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.bar(x_axis, y_axis, color = "g", label = "Boston")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis, y_axis, color = "m", label = "Boston")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "skyblue", alpha = .8, linewidth = 2, edgecolors = "k", label = "Boston", s = [5*i for i in y_axis])
ax.invert_yaxis()
ax.set_xlim(0, 50)
ax.legend()
```

```python
#plt.pie(y_axis, labels = x_axis)
#plt.show()
explode = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize = (8,8))
plt.pie(y_axis, labels = x_axis, explode = explode, autopct="%.1f%%")
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,8))
explode = (0,0,.2,0,0,0,.2,0,0,0,0,0)
ax.pie(y_axis, labels = x_axis, autopct = "%.1f%%", explode = explode, shadow = True, startangle = 100, counterclock = False, colors = colors)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import statistics

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get sd for y vals
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr = stdev, capsize = 3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step = 5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator as minor
#Incraese plot size 
fig, ax = plt.subplots(figsize = (8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, 5.0))

#Create minor ticks inc = 1
ax.xaxis.set_minor_locator(minor(1))
plt.show()
```
