```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Load csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)")
plt.show()
```

```python
#Set x-axis and ticks
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [val for val in x_axis]

#Plot
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x = "Month", y = "Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)", kind = "bar")
plt.show()
```

```python
import statistics as stat
stdev = stat.stdev(pyber_ride_df["Avg. Fare ($USD)"])
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)", kind = "bar", yerr = stdev, capsize = 3, color = "skyblue", rot = 0)
#rot param from https://stackoverflow.com/questions/32244019/how-to-rotate-x-axis-tick-labels-in-pandas-barplot
plt.yticks(np.arange(0, 51, 5))
```
